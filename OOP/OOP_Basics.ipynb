```python
class Person:
  #static attributes
  count = 0

  #instance attributes
  def __init__(self, name, address, number, email):
    self.name = name
    self.address = address
    self.number = number
    self.__email = email
    Person.count += 1

  # def __str__(self):
  #   return str(self.__dict__)

  #using Decorators - - affects the method directly below
  @property
  def email(self):
    return self.__email

  @email.setter
  def email(self, email):
    self.__email = email

person1 = Person("Adam", "Baker Street", "0123456789", "adam@email.com")

#printing static attributes
print(Person.count)
print(person1.count)

# print(person1.__email) - not accessible
print(person1.email)
person1.email = "newemail@email.com"

print(person1.email)
```

```python
class BankAccount():
  #CONSTANT static attribute
  MIN_BALANCE = 100

  #instance attributes
  def __init__(self, owner, balance):
    if balance < BankAccount.MIN_BALANCE:
      raise ValueError(f"balance should be higher than ${BankAccount.MIN_BALANCE} to be able to create an account")
    else:
      self.owner = owner
      self.balance = balance

  #instance methods
  def deposit(self, amount):
    if self._is_valid_amount(amount):
      self.balance = self.balance + amount
      print(f"balance after deposti: ${self.balance}")
    else:
      print("invalid amount")

  def withdraw(self, amount):
    if self._is_valid_amount(amount) and self.balance - amount > BankAccount.MIN_BALANCE:
      self.balance = self.balance - amount
      print(f"balance after withdrawal: ${self.balance}")
    else:
      print(f"can't withdraw ${amount} as account balance will go bleow limit as a result of this withdrawal")

  #static method - class use only
  @staticmethod
  def updateMinBalance(amount):
    if BankAccount._is_valid_amount(amount):
      BankAccount.MIN_BALANCE = amount

    print(BankAccount.MIN_BALANCE)

  #protected method, changed to staticmethod as it doesn't use 'self'
  @staticmethod
  def _is_valid_amount(amount):
    return isinstance(amount, (float, int))

try:
  acc = BankAccount(person1, 500)
  acc.deposit(100)
  acc.withdraw(300)
  BankAccount.updateMinBalance(200)
except ValueError as e:
  print(e)
```

```python
#Inheritance & Polymorphism

class Vehicle():
  def __init__(self, genus, wheels, color):
    self.genus = genus
    self.wheels = wheels
    self.color = color

  def start(self):
    print("vehicle engine started")

  def stop(self):
    print("vehicle engine stopped")

class Car(Vehicle):
  def __init__(self, genus, wheels, color, brand, doors):
    #super class initiation
    super().__init__(genus, wheels, color)
    self.brand = brand
    self.doors = doors

class Bike(Vehicle):
  def __init__(self, genus, wheels, color, brand):
    #super class initiation
    super().__init__(genus, wheels, color)
    self.brand = brand

  #Polymorphism
  def start(self):
    print("bike engine started")
  def stop(self):
    print("bike engine stopped")

car = Car("four-wheeler", 4, "red", "Toyota", 4)
car.start()
print(car.genus)
print(car.wheels)
print(car.__dict__)

vehicles = [
    Car("four-wheeler", 4, "red", "Toyota", 4),
    Bike("two-wheeler", 2, "black", "Harley Davidson")
]

for vehicle in vehicles:
  if isinstance(vehicle, Vehicle):
    print(vehicle.__dict__)
    vehicle.start()
    vehicle.stop()
```
